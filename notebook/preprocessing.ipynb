```python
import pandas as pd
```

```python
# load sample data

sample_path = '../data/sample/50Cent_CandyShop.rap'
sample = pd.read_table(sample_path)
```

```python
sample.head(10)
```

```python
# check the rows with na 

sample[sample.isnull().any(axis=1)]
#sample.shape
# na rows are metadata rows. We are safe to proceed.
```

```python
# save only the relevant metadata : release date, song title, artist
#title = sample['sample']
title = sample['**recip'][sample['**recip'].apply(lambda x: 'OTL:' in x)].values[0]
artist = sample['**recip'][sample['**recip'].apply(lambda x: 'COC:' in x)].values[0]
release = sample['**recip'][sample['**recip'].apply(lambda x: 'CDT:' in x)].values[0]
tempo = sample.iloc[1]['**recip']
meter = sample.iloc[0]['**recip']
print(title, artist, release, tempo, meter)
```

```python
# count the measure indicators and verse indicators
measures = sample['**recip'][sample['**recip'].apply(lambda x: '=' in x)]
verses = sample['**recip'][sample['**recip'].apply(lambda x: 'Verse' in x)]

#measures.append(verses).index
```

```python
sample['**measure'] = 'None'
sample['**verse'] = 'None'
#sample['**measuar'] = sample

for i in measures.index:
    sample.at[i+1,'**measure'] =  measures.loc[i]
 
for i in verses.index:
    sample.at[i+1,'**verse'] =  verses.loc[i]
 
    
sample
```

```python
sample =  sample.drop([0,1,3])
sample = sample.dropna(axis=0)
sample = sample.drop(measures.index, axis=0)
sample = sample.drop(verses.index, axis=0)
sample#.shape
```

```python
def cast_break(x):
    res = x
    if x in ['3','4']:
        res = int(x)
    return res

sample['**break'] = sample['**break'].apply(cast_break)
breaks_idx = sample[sample['**break'].apply(lambda x: x in [3,4])].index
# add closing break point

sample_flowwise = pd.DataFrame(columns=['recip','stress','lyric'])


for i in range(breaks_idx.shape[0]-1):
    (start, end) = (breaks_idx[i], breaks_idx[i+1])
    snippet = sample.iloc[start:end]
    sample_flowwise.at[i,'recip'] = snippet['**recip'].values
    sample_flowwise.at[i,'lyric'] = ' '.join(snippet['**lyrics'].values)
```
